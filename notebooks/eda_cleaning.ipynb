# EDA Cleaning Notebook — Customer Risk Analysis

```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv("../data/raw_customer_risk.csv")
df.head()
```

```
df.shape
df.info()
df.describe()
df.isnull().sum()
df.duplicated().sum()
df.columns
```

```
df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]
df["claim_date"] = pd.to_datetime(df["claim_date"], errors="coerce")
for col in ["tenure_years", "annual_premium", "claim_amount"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")
df = df.drop_duplicates()
df.head()
```

```
customer_summary = df.groupby("customer_id").agg(total_claims=("claim_id","count"), total_loss=("claim_amount","sum"), avg_loss=("claim_amount","mean")).reset_index()
customer_summary.head()
```

```
claim_threshold = 4
loss_threshold = 15000
customer_summary["risk_category"] = customer_summary.apply(lambda r: "High-risk" if r.total_claims >= claim_threshold or r.total_loss >= loss_threshold else ("Medium-risk" if r.total_claims >= 2 or r.total_loss >= 8000 else "Low-risk"), axis=1)
customer_summary["risk_category"].value_counts()
```
